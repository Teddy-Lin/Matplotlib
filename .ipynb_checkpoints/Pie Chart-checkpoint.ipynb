```python
# pie chart doesn't work well with many categories, bar chart works better
from matplotlib import pyplot as plt
```

```python
plt.style.use("ggplot")
```

```python
slices = [59219, 55466, 47544, 36443, 35917] # doesn't have to add up to 100
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0] # emphasize something
# 0.1 means 10% of the radius from the center
colors = ["#008fd5", "#fc4f30", "#e5ae37", "green", "pink"] # remember to put in quotes
plt.pie(slices, 
        labels = labels, 
        colors = colors, 
        explode = explode, 
        wedgeprops = {"edgecolor" : "black"}, 
        shadow = True, 
        autopct = "%1.1f%%",
        startangle = 90)

plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```
